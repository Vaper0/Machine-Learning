Список підключених бібліотек

```
import numpy as ny
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans, MeanShift, DBSCAN
import statistics
```

Створюєм датасет


```
X1, Y1 = datasets.make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=3)
#X3, Y3 = datasets.make_moons(n_samples=200, noise=0.3, random_state=42) - тестовий датасет, в розбір не війшов
plt.scatter(X1[:,0], X1[:,1])
plt.show()
```

Застосовуємо різні методи кластерного аналізу

```
#KMeans (використано для трьох кластерів, хоча датасет генерував з двома)
km = KMeans(n_clusters=3).fit(X1)
centroids = km.cluster_centers_
labels = km.labels_
plt.scatter(X1[:,0], X1[:,1], c= km.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

```
#MeanShift
ms = MeanShift().fit(X1)
centroids = ms.cluster_centers_
labels = ms.labels_
plt.scatter(X1[:,0], X1[:,1], c= km.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

```
#DBSCAN
db = DBSCAN(eps=0.33, min_samples=20).fit(X1)
plt.scatter(X1[:,0], X1[:,1], c= db.labels_.astype(float), s=50, alpha=0.5)
plt.show()
```

Як видно після першого блоку, кожен метод працює по своєму, потребує індивідуального розуміння як і коли його застосовувати. Це в свою чергу говорить проте що найкращо методу серед них нема все залежить від дата сету. Нижче наведу приклад з іншим датасетом.

```
X2, Y2 = datasets.make_blobs(n_samples=200, centers=3, cluster_std=0.60, random_state=7)
plt.scatter(X2[:,0], X2[:,1])
plt.show()
```

```
#KMeans
km = KMeans(n_clusters=3).fit(X2)
centroids = km.cluster_centers_
labels = km.labels_
plt.scatter(X2[:,0], X2[:,1], c= km.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()

#MeanShift
ms = MeanShift().fit(X2)
centroids = ms.cluster_centers_
labels = ms.labels_
plt.scatter(X2[:,0], X2[:,1], c= km.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()

#DBSCAN
db = DBSCAN(eps=0.80, min_samples=10).fit(X2)
plt.scatter(X2[:,0], X2[:,1], c= db.labels_.astype(float), s=50, alpha=0.5)
plt.show()
```

Статистичний аналіз за допомогою бібліотеки statistics. Його я проводитиму на другому датасеті(бо зручний) з розподілом KMeans.

```
print(statistics.mean(X2[Y2 == 0, 0]))
print(statistics.mean(X2[Y2 == 1, 0]))
```
