#Бібліотеки

```python
from random import randrange
import numpy as np
from time import time
import unittest
```

#Функції

Генерація пустих матриці та вектора

```python
def zeros_matrix(rows, cols):
    M = []
    while len(M) < rows:
        M.append([])
        while len(M[-1]) < cols:
            M[-1].append(0.0)
 
    return M
```

```python
def zero_vector(len):
    M = []
    for i in range(len):
        M.append(0)
    return M
```

Генерація/Заповнення матриці, вектора рандомними числами

```python
def random_matrix(rows, cols):
    M = []
    while len(M) < rows:
        M.append([])
        while len(M[-1]) < cols:
            M[-1].append(randrange(10))
 
    return M
```

```python
def random_vector(len):
    M = []
    for i in range(len):
        M.append(randrange(10))
    return M
```

Множення матриці на матрицю

```python
def matrix_multiply(A, B):
    rowsA = len(A)
    colsA = len(A[0])
    rowsB = len(B)
    colsB = len(B[0])
    if colsA != rowsB:
        raise ArithmeticError('Кількість колонок не співпадає з рядками')
    C = zeros_matrix(rowsA, colsB)
    for i in range(rowsA):
        for j in range(colsB):
            total = 0
            for ii in range(colsA):
                total += A[i][ii] * B[ii][j]
            C[i][j] = total
 
    return C
```

Множення матриці на вектор

```python
def matrix_vector(A, N):
    rowsA = len(A)
    colsA = len(A[0])
    leng = len(N)
    L = zero_vector(leng)
    for i in range(colsA):
        total = 0
        for j in range(rowsA):
            total += A[i][j] * N[j]
        L[i] = total
    return L
```

Множення вектора на матрицю

```python
def vector_matrix(A, N):
    rowsA = len(A[0])
    colsA = len(A)
    leng = len(N)
    L = zero_vector(leng)
    for i in range(colsA):
        for j in range(rowsA):
            L[i] += A[j][i] * N[j]
    return L
```

Множення вектора на вектор

```python
def vector_vector(A,B):
    leng = len(A)
    L = 0
    for i in range(leng):
        L = L + A[i]*B[i]
    return L
```

#Unittest

```python
class MyTestCase(unittest.TestCase):
    
    def test_matrix_multiply(self):
        a = np.array([[1, 2], [5, 6]])
        b = np.array([[2, 4], [1, 5]])
        result = matrix_multiply(a,b)
        need = np.array(np.dot(a,b))
        for i in range(2):
          for j in range(2):
              self.assertEqual(result[i][j],need[i][j])

    def test_matrix_vector(self):
        a = np.array([[1, 2], [5, 6]])
        b = np.array([2, 4])
        result = matrix_vector(a,b)
        need = np.array(np.dot(a,b))
        for i in range(2):
              self.assertEqual(result[i],need[i])

    def test_vector_matrix(self):
        a = np.array([[1, 2], [5, 6]])
        b = np.array([2, 4])
        result = vector_matrix(a,b)
        need = np.array(np.dot(b,a))
        for i in range(2):
              self.assertEqual(result[i],need[i])

    def test_vector_vector(self):
        a = np.array([1, 2])
        b = np.array([2, 4])
        result = vector_vector(a,b)
        need = np.dot(a,b)
        self.assertEqual(result,need)


if __name__ == '__main__':
   unittest.main(argv=['first-arg-is-ignored'], exit=False)
```

#Порівняння роботи Python vs Python + NumPy

Ініціалізація всіх зміних

```python
n = 100
m = 100
A = random_matrix(n,m)
B = random_matrix(n,m)
N = random_vector(n)
L = random_vector(m)
```

Множення матриць

```python
t0 = time()
C = matrix_multiply(A,B)
t1 = time()


#Numpy
t2 = time()
D = np.dot(A,B)
t3 = time()


print("Множення матриць без Numpy:",t1-t0,"\n",C,"\n")
print("Множення матриць з Numpy:",t3-t2,"\n",D)
```

Множення матриці на вектор

```python
t0 = time()
E = matrix_vector(A,N)
t1 = time()


#Numpy
t2 = time()
F = np.dot(A,N)
t3 = time()

print("Множення матриці на вектор без Numpy:",t1-t0,"\n",E,"\n")
print("Множення матриці на вектор з Numpy:",t3-t2,"\n",F)
```

Множення вектора на матрицю

```python
t0 = time()
G = vector_matrix(A,N)
t1 = time()

#Numpy
t0 = time()
H = np.dot(N,A)
t1 = time()

print("Множення вектора на матрицю без Numpy:",t1-t0,"\n",G,"\n")
print("Множення вектора на матрицю з Numpy:",t3-t2,"\n",H)
```
